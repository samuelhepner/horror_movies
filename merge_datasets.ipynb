## Python notebook to merge the two datasets together

```python
# Let's import some libraries
import pandas as pd
```

```python
# Loading our datasets
movie_info = pd.read_csv("imdb_movies/IMDb_movies.csv")
movie_ratings = pd.read_csv("imdb_ratings/IMDb_ratings.csv")
```

```python
movie_ratings.head()
```

```python
# Let's take a look at the columns for movie_ratings
movie_ratings.columns
```

```python
"""For next steps, I want to dive deeper into the data and get a better understanding of what columns votes_10 - votes_1 are and if they
    would be helpful for this project.
    
    After that, I would like to join the two csv files and make several visualizations."""
```

```python
# I'm still not sure what the columns vote_10 - vote_1 are counting
# but I'm going to go ahead and join the datasets together and do more research later!

imdb_dataset = pd.merge(movie_info, movie_ratings, on='imdb_title_id', how='outer')
# Check to make sure all the columns are there
imdb_dataset.head()
```

```python
# Let's save this merged dataset as a csv so we don't have to run the lines above again
imdb_dataset.to_csv('imdb_dataset.csv')
```

### Taking a look at the official imdb tsv datasets

```python
# Let's import some libraries
import pandas as pd
```

```python
# Let's try reading a tsv file
akas = pd.read_csv('title_akas.tsv', sep='\t')
basics = pd.read_csv('title_basics.tsv', sep='\t')
# crew = pd.read_csv('title_crew.tsv', sep='\t')
ratings = pd.read_csv('title_ratings.tsv', sep='\t')
```

```python
akas.head()
```

```python
renamed_akas = akas.rename(columns={"titleId": "tconst"})
```

```python
renamed_akas.head()
```

```python
basics.head()
```

```python
ratings.head()
```

```python
# Merge all datasets together.
# For now, I'll merge all but the crew for now.

akas_and_basics = pd.merge(renamed_akas, basics, on="tconst", how="outer")
akas_basics_ratings = pd.merge(akas_and_basics, ratings, on="tconst", how="outer")
```

```python
# Let's save it as a tsv so we don't have to do the merging again

akas_basics_ratings.to_csv("imdb_datasets.tsv", sep="\t")
```
